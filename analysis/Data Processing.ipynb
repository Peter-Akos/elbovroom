```python
import pandas as pd
```

```python
df = pd.read_csv("../data/vigyazz-3000-3049.csv")
```

```python
import re

def extract_date(details):
    match = re.search(r'EZ (\d{2}/\d{4})', details)
    return match.group(1) if match else None

def extract_kilometers(details):
    match = re.search(r'(\d{1,3}(?:\.\d{3})*) km', details)
    return int(match.group(1).replace('.', '')) if match else None

def extract_kw(details):
    match = re.search(r'(\d{2,3}) kW', details)
    return int(match.group(1)) if match else None

def extract_ps(details):
    match = re.search(r'\((\d{2,3}) PS\)', details)
    return int(match.group(1)) if match else None

def extract_fuel(details):
    match = re.search(r'(Benzin|Diesel|Hybrid|Elektro|Gas)', details, re.IGNORECASE)
    return match.group(1) if match else None

def extract_accident(details):
    if 'Unfallfrei' in details:
        return False
    elif 'Unfall' in details or 'Reparierter Unfallschaden' in details:
        return True
    return None
```

```python
def process_dataframe(df):
    df['date'] = df['details'].apply(extract_date)
    df['kilometers'] = df['details'].apply(extract_kilometers)
    df['kw'] = df['details'].apply(extract_kw)
    df['ps'] = df['details'].apply(extract_ps)
    df['fuel'] = df['details'].apply(extract_fuel)
    df['accident'] = df['details'].apply(extract_accident)
    return df
    
```

```python
process_dataframe(df)
```
